## With cut-off of significant slopes

```python
## === Find the p-value === ##

summary_measure_array = summary_measures.fit_shm(normalised_flattened_masked_dwi, shm_degree=2, bval=bvalues, bvec=bvecs)

number_voxels = summary_measure_array.shape[0]
y = summary_measure_array

## y = X @ beta (voxels x sm  = voxels x change axes @ change axes x sm)

axes_of_change = [x for x in dictionary_of_axes.keys()]

X = np.ones(number_voxels)

# generate design matrix (X)
for axe in axes_of_change:
    axis_of_change = dictionary_of_axes[axe]
    X = np.column_stack((X, axis_of_change - axis_of_change.mean()))
```

```python
import statsmodels.api as sms

p_array_nissl = np.ones([1,5])
p_array_myelin = np.ones([1,5])

for i, s in enumerate(sm_names):
    mod = sms.OLS(y[:,i],X)
    fii = mod.fit()
    
    p_array_nissl[0,i] = fii.summary2().tables[1]['P>|t|'][1]
    p_array_myelin[0,i] = fii.summary2().tables[1]['P>|t|'][2]
```

```python
p_mask_nissl = p_array_nissl>=0.05, 

p_mask_myelin = p_array_myelin>=0.05 #use this to mask the deltas and set them to arbitrarily small number. 
```

```python
## mask out the insignificant slopes

baseline = dictionary_of_betas["baseline"].mean(axis=0)[np.newaxis,:]
delta_nissl = dictionary_of_betas["nissl"].mean(axis=0)[np.newaxis,:]
delta_myelin = dictionary_of_betas["myelin"].mean(axis=0)[np.newaxis,:]
noise_cov_nissl = dictionary_of_noise_covar["nissl"]
noise_cov_myelin = dictionary_of_noise_covar["myelin"]


#delta_nissl[p_mask_nissl] = 1e-5
#delta_myelin[p_mask_myelin] = 1e-5

#minimum_val = np.abs(noise_cov_nissl).min()

#noise_cov_nissl[2,2] = minimum_val
#noise_cov_nissl[-1,-1] = minimum_val

#noise_cov_nissl[:,2] = minimum_val
#noise_cov_nissl[2,:] = minimum_val
#noise_cov_nissl[:,-1] = minimum_val
#noise_cov_nissl[-1,:] = minimum_val

prob_nissl, _, _, _  = ch_mdl.infer(baseline,
                              delta_nissl,
                              noise_cov_nissl,
                              parallel=False)

prob_myelin, _, _, _  = ch_mdl.infer(baseline,
                              delta_myelin,
                              noise_cov_myelin,
                              parallel=False)
```

## Computing the covariance of betas from the GLM (the square of the standard errors of the models' coefficients)

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()

covar_nissl = np.zeros((len(sm_names), len(sm_names)))
covar_myelin = np.zeros_like(covar_nissl)

ols_delta_nissl = np.zeros((1, len(sm_names)))
ols_delta_myelin = np.zeros((1, len(sm_names)))
ols_baseline = np.zeros_like(delta_nissl)

for i, sum_meas in enumerate(sm_names):

    model = regr.fit(X, y[:,i])
    MSE = np.mean((y[:,i] - regr.predict(X))**2)
    var_est = MSE * np.diag(np.linalg.pinv(np.dot(X.T,X)))
    covar_nissl[i,i] = var_est[1] 
    covar_myelin[i,i] = var_est[2] 
    ols_delta_nissl[0,i] = model.coef_[1] * 0.4
    ols_delta_myelin[0,i] = model.coef_[2] * 0.4 
    ols_baseline[0,i] = model.intercept_
    
```
