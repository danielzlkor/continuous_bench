### GLM

```python
#summary_measure_array = summary_measures.fit_shm(normalised_flattened_masked_dwi, shm_degree=2, bval=bvalues, bvec=bvecs)

summary_measure_array = np.copy(normalised_dwi_sm)
number_voxels = summary_measure_array.shape[0]
y = summary_measure_array

## y = X @ beta (voxels x sm  = voxels x change axes @ change axes x sm)

axes_of_change = [x for x in dictionary_of_axes.keys()]

X = np.ones(number_voxels)

# generate design matrix (X)
for axe in axes_of_change:
    axis_of_change = dictionary_of_axes[axe]
    X = np.column_stack((X, axis_of_change - axis_of_change.mean()))
```

### All samples used GLM

```python
## All samples method

X_least_squares = np.linalg.inv(X.T @ X) @ X.T
betas_bar = X_least_squares @ y

mean_idx = np.array([[1, 0, 0]])
nissl_idx = np.array([[0, 1, 0]])
myelin_idx = np.array([[0, 0, 1]])

#mean_idx = np.array([[1, 0]])
#nissl_idx = np.array([[0, 1]])


res = (y - X @ betas_bar)

cov_nissl = (res.T @ np.diag(np.diag((X_least_squares).T @ nissl_idx.T @ nissl_idx @ X_least_squares)) @ res) * effect_size_nissl**2
beta_nissl = (nissl_idx @ betas_bar) * effect_size_nissl

cov_myelin = (res.T @ np.diag(np.diag((X_least_squares).T @ myelin_idx.T @ myelin_idx @ X_least_squares)) @ res) * effect_size_myelin**2
beta_myelin = (myelin_idx @ betas_bar) * effect_size_myelin

beta_baseline = mean_idx @ betas_bar
```

## With cut-off significant slopes

```python
## === Find the p-value === ##

summary_measure_array = summary_measures.fit_shm(normalised_flattened_masked_dwi, shm_degree=2, bval=bvalues, bvec=bvecs)

number_voxels = summary_measure_array.shape[0]
y = summary_measure_array

## y = X @ beta (voxels x sm  = voxels x change axes @ change axes x sm)

axes_of_change = [x for x in dictionary_of_axes.keys()]

X = np.ones(number_voxels)

# generate design matrix (X)
for axe in axes_of_change:
    axis_of_change = dictionary_of_axes[axe]
    X = np.column_stack((X, axis_of_change - axis_of_change.mean()))
```

```python
import statsmodels.api as sms

p_array_nissl = np.ones([1,5])
p_array_myelin = np.ones([1,5])

for i, s in enumerate(sm_names):
    mod = sms.OLS(y[:,i],X)
    fii = mod.fit()
    
    p_array_nissl[0,i] = fii.summary2().tables[1]['P>|t|'][1]
    p_array_myelin[0,i] = fii.summary2().tables[1]['P>|t|'][2]
```

```python
p_mask_nissl = p_array_nissl>=0.05, 

p_mask_myelin = p_array_myelin>=0.05 #use this to mask the deltas and set them to arbitrarily small number. 

p_array_nissl, p_array_myelin
```
