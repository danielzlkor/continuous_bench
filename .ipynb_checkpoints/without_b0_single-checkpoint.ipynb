```python
def estimated_change_vectors(ch_mdl, data):
        """
            Returns the change vectors given the baseline measurement (mu, sigma)
        :param data: un-normalized baseline measurements.(n, d) n= number of samples, d= number of measurements
        :return: mu (n, m, d) , sigma_p(n, m, d,d) m=number of change vecs.
        """
        data = np.atleast_2d(data)
        n_samples, d = data.shape
        assert d == len(ch_mdl.summary_names)
        n_models = len(ch_mdl.models)

        mu = np.zeros((n_samples, n_models, d))
        sigma_p = np.zeros((n_samples, n_models, d, d))

        for i, y in enumerate(data):
            for j, mdl in enumerate(ch_mdl.models):
                #print(mdl.distribution(y))
        
                if j==0:
                    mu[i, j] = ch_mdl.models[j].distribution(data)[0][:,1:]
                    sigma_p[i,j] = ch_mdl.models[j].distribution(data)[1][:,1:,1:]
                else:
                    mu[i, j], sigma_p[i, j] = mdl.distribution(y)
        
        return np.squeeze(mu), np.squeeze(sigma_p)
```

```python
def overlap_full_change(sm_names, amount_dict, inf_change_dict,
                        mus, sigmas,
                        param,
                       together=False,
                       ylim=None):
    
    s = sigmas[inf_change_dict[param]][0,...]
    dv = amount_dict[param]
    s_data = covar_dict[param] + np.eye(5)*reg
    
    if together: 
        
            continuous_change_glm.plot_changes(sm_names, beta_dict[param])
        
            continuous_change_glm.plot_changes(sm_names,
                                   (amount_dict[param]*mus[inf_change_dict[param]]),
                                   error=np.sqrt(np.diag(s_data + s*dv**2)),
                                  color="red")


   
    else:
        
        continuous_change_glm.plot_changes(sm_names, beta_dict[param], error=np.sqrt(np.diag(s_data)))

        continuous_change_glm.plot_changes(sm_names,
                                       (amount_dict[param]*mus[inf_change_dict[param]]),
                                       error=np.sqrt(np.diag(s*dv**2)),
                                      color="red")
        

        
    plt.title(param, fontsize=15)
    plt.xticks(fontsize=4)
    
    if ylim is not None:
        
        plt.ylim(ylim)


    
    
    
    
```

## Things to consider

- How much the l2 term is considered is dependent on the ratio between the mean change (i.e. change in summary measure due to change in variable) and the total variance (variance in summary measure due to a change in parameter + variance of a change in summary due to a change in continuous variable).

- The higher the variance, the less likely the summary measure is considered as part of the inference (i.e. low probability) (i.e. increase the sigma here --> 1/sigma goes to 0, so that probability is low) 



- Alternatively, one way to tell how much it relies on a summary measure to make a prediction is to just look at the noise covariance.
- Take the noise covariance matrix, and perform an eigenvalue decomposition. Then, take the lambdas (min)/(max) and see the big difference. Check the eigenvectors to see which summary measures correspond with which principle direction. Do this for all continuous variables. The ratio for ODIs should be BIGGER. 

- Visualisation (to make sure that your story is correct) --> Figure 11 of Hossein's paper. Multiply the patterns of change with amounts of change.

```python

def find_eigens(matrix):
    # Check if the input is a 5x5 matrix
    
    # Calculate the eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(matrix)

    # Sort the eigenvalues, and find the index of the 2nd smallest (2nd minimum)
    sorted_indices = np.argsort(eigenvalues)
    second_smallest_index = sorted_indices[1]  # 2nd element after sorting (since it starts at 0)
    max_eigenvalue_index = sorted_indices[-1]  # The last element is the largest

    # Find the corresponding eigenvectors
    second_smallest_eigenvector = eigenvectors[:, second_smallest_index]
    max_eigenvector = eigenvectors[:, max_eigenvalue_index]

    # Calculate the ratio of the 2nd smallest to the maximum eigenvalues
    eigenvalue_ratio = eigenvalues[second_smallest_index] / eigenvalues[max_eigenvalue_index]

    print(f"Eigenvalues: {eigenvalues}")
    print(f"Eigenvectors: \n{eigenvectors}")
    print(f"Ratio of 2nd smallest to max eigenvalues: {eigenvalue_ratio}")
    print(f"Eigenvector corresponding to the 2nd smallest eigenvalue: {second_smallest_eigenvector}")
    print(f"Eigenvector corresponding to the max eigenvalue: {max_eigenvector}")

    return eigenvalues, eigenvectors, eigenvalue_ratio, second_smallest_eigenvector, max_eigenvector

```
