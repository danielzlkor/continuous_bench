```python
## === Signal bookkeeping ===

import csv

def read_csv(filename):
    with open(filename, newline='') as f_input:
        return [list(map(float, row)) for row in csv.reader(f_input)]
    
def signal_invivo_loader(signal_csv):
    
    if len(read_csv(signal_csv)) == 0:
        print("{} is empty".format(signal_csv))
        return

    signal = np.array(read_csv(signal_csv)[0])
    
    b0 = signal[0]

    signal_attenuation = signal/b0
    
    signal_attenuation = signal_attenuation.reshape(1,signal.shape[0])
    
    return signal_attenuation

import re

def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]
```

```python
import matplotlib as mpl
import numpy as np

from re import split

def default_template():
    mpl.rcParams["figure.facecolor"] = "FFFFFF"
    mpl.rcParams["axes.facecolor"] = "FFFFFF"
    mpl.rcParams['font.family'] = 'Arial'
    mpl.rcParams['axes.labelcolor']="000000"
    mpl.rcParams['axes.edgecolor']="000000"


    mpl.rcParams['ytick.color']="000000"
    mpl.rcParams['xtick.color']="000000"
    mpl.rcParams['ytick.labelsize']=50
    mpl.rcParams['xtick.labelsize']=50
    

def plot_probs(probs, free_params, title, fontsize=25):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    
    colors=['b', 'g', 'r', 'm', 'c', 'y']
    
    ax.set_ylim(0,1)
    
    colors={"None": "c",
            "s_in": 'b',
            "s_ex": 'k',
            "s_dot": "brown",
            "s_sphere": "orchid",
            "d_sphere": 'purple',
            "rad_sphere": 'gold', 
            "odi" : 'r',
            "d_a_in":'g',
            "d_a_ex" : 'm',
            "tortuosity": 'y'}
    
    notation={"None": "$None$",
              "s_in": "$f_{in}$",
              "s_ex": "$f_{ex}$",
              "s_dot": "$f_{dot}$",
              "s_sphere": "$f_{sphere}$",
            "d_sphere": '$D_{sph}$',
            "rad_sphere": '$R_{sph}$', 
            "odi" : "$ODI$",
            "d_a_in":"$D_{a,in}$",
            "d_a_ex" : "$D_{a,ec}$",
            "tortuosity": r"$\tau$"}
    
    probs = np.squeeze(probs)
    
    for i, x in enumerate(free_params):
        
        ax.bar(notation[x], probs[i], color = colors[x], width = 0.80)
    
    
    ax.set_title(title, fontsize=30)
    ax.spines["bottom"].set_visible(True)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    
```

```python
def continuous_change(summary_measure_array, dictionary_of_axes):
    
    ## === Run GLM to get y, delta_y and sn ===
    number_voxels = summary_measure_array.shape[0]
    mean_data = np.repeat(summary_measure_array.mean(axis=0)[np.newaxis,...], repeats=number_voxels, axis=0)
    y = summary_measure_array 

    ## y = X @ beta (voxels x sm  = voxels x change axes @ change axes x sm)
    
    axes_of_change = [x for x in dictionary_of_axes.keys()]
    
    X  = np.ones(number_voxels)
    
    # generate design matrix (X) 
    for axe in axes_of_change:
        
        axis_of_change = dictionary_of_axes[axe]
        X = np.column_stack((X, axis_of_change-axis_of_change.mean()))
        
    # regression 
    
    betas = np.linalg.pinv(X) @ (y)
    
    # designate directions of change
    
    delta_ys = {}

    for idx,axe in enumerate(axes_of_change): 

        delta_ys[axe] = betas[idx+1,:][np.newaxis,...]
        
    
    # designate baseline
    baseline_y = betas[0, :]
    baseline_y = baseline_y[np.newaxis,...]

    
    return baseline_y, delta_ys
```

```python
def fit_group_shm_with_noise_and_continuous_noise_covariance(signals, substrates,  dictionary_of_axes, sm_names,  bvalues, gradient_directions, shm_degree, SNR, iterations=100, noise_type="gaussian"):
    
    
    noisy_signal_dictionary = {}
    
    
    for j, substrate in enumerate(substrates): #iterate over substrates to generate SM

        
        signal = signals[substrate]
        
        # add noise to signal
        noisy_signal = add_noise(signal, noise_type=noise_type, SNR=SNR, S0=1)
        
        noisy_signal_dictionary[substrate] = noisy_signal
        
        sm = summary_measures.fit_shm(noisy_signal, bvalues, gradient_directions, shm_degree=shm_degree)
            
        #concatenate 
        
        if j == 0:
            
            grp_sm = sm
        
        else:
            
            grp_sm = np.concatenate((grp_sm, sm), axis=0)
            
    
    #for noise covariance
    dictionary_of_noise_covar = estimate_continuous_noise_covariance(noisy_signal_dictionary, substrates, dictionary_of_axes, sm_names, iterations, bvalues, gradient_directions, SNR, sm="shm", shm_degree=2, noise_type=noise_type) 

    
    return grp_sm, dictionary_of_noise_covar, noisy_signal_dictionary
```

```python
def add_noise(data, noise_type='gaussian', SNR=1, S0=1):
    """Add noise to signal

    For Gaussian :
            Snoise = S0 + N(0,sigma)
    For Rician :
            Snoise = sqrt( Real^2 + Imag^2  )
            where Real = S0+N(0,sigma) and Imag=N(0,sigma)
    In both case SNR is defined as SNR = S0/sigma

    Args:
        data : array-like
        noise_type : str
                     accepted values are 'gaussian', 'rician', or 'none'
        SNR : float
        S0  : float
    """
    if noise_type not in ['none','gaussian','rician']:
        raise(Exception(f'Unknown noiise type {noise_type}'))
    if noise_type == 'none':
        return data+0.
    else:
        sigma = S0 / SNR
        if noise_type == 'gaussian':
            noise = np.random.normal(loc=0.0, scale=sigma, size=data.shape)
            return data + noise
        elif noise_type == 'rician':
            noise_r = np.random.normal(loc=0.0, scale=sigma, size=data.shape)
            noise_i = np.random.normal(loc=0.0, scale=sigma, size=data.shape)
            return np.sqrt( (data+noise_r)**2 + noise_i**2 )


def fit_shm_with_noise(signal, bvalues, gradient_directions, shm_degree, SNR, noise_type="gaussian"):
    
    noisy_signal = add_noise(signal, noise_type=noise_type, SNR=SNR, S0=1)
    
    sm = summary_measures.fit_shm(noisy_signal, bvalues, gradient_directions, shm_degree=shm_degree) 
    
    return sm

def estimate_continuous_noise_covariance(signals, substrates, dictionary_of_axes, sm_names, iterations, bvalues, gradient_directions, SNR, sm="shm", shm_degree=2, noise_type="gaussian"):
    
    ## This is for a single signal, with no samples to draw the covariance from
    
    dictionary_of_betas = {}
    dictionary_of_noise_covar = {}
    
    # get the noise covariance for each beta 
    
    axes_of_change = [x for x in dictionary_of_axes.keys()]
    
    for axe in axes_of_change: 
        dictionary_of_betas[axe] = np.zeros((iterations, len(sm_names)))
        dictionary_of_noise_covar[axe] = np.zeros((iterations, len(sm_names)))

    for i in range(iterations):
        
        if sm=="shm":
            
            for j, substrate in enumerate(substrates):
                
                noisy_sm = fit_shm_with_noise(signals[substrate], bvalues, gradient_directions, shm_degree, SNR, noise_type=noise_type)
                
                #concatenate 

                if j == 0:

                    grp_sm = noisy_sm

                else:

                    grp_sm = np.concatenate((grp_sm, noisy_sm), axis=0)
                    

            #infer changes across all substrate
            baseline_y, delta_ys = continuous_change(grp_sm, dictionary_of_axes)
                                  
            for axe in axes_of_change: #iterate over all change axes and compute beta for deriving covariance
        
                dictionary_of_betas[axe][i,:] = delta_ys[axe]
            

    
    for axe in axes_of_change: 
        
        #print(dictionary_of_betas[axe].shape)
        
        dictionary_of_noise_covar[axe] = np.cov(dictionary_of_betas[axe].T)
        
            
    
    return dictionary_of_noise_covar
```

```python

axon = [3,4]
sphere = [0, 1, 2, 3, 4, 5, 6, 7]

#axon = axon_list
#sphere = sphere_list


substrates = []
dictionary_of_axes = {}

existing_axons = []
existing_spheres = []

for x in axon:
    for y in sphere:
        
        substrate = "axon_{}_sphere_{}".format(x,y)
        
        if substrate not in dict_of_signals.keys():
            print("{} does not exist".format(substrate))
            
        
        else:
            substrates.append(substrate)
            existing_axons.append(x)
            existing_spheres.append(y)

        
# set up dictionaries
dictionary_of_axes["axon"] = np.array(existing_axons)
dictionary_of_axes["sphere"] = np.array(existing_spheres)

dictionary_of_axes
```

```python
shm_degree = 2
SNR=100

iterations = 10

dictionary_of_probs_params = {}

noise_type = "gaussian"

axes_of_change = [x for x in dictionary_of_axes.keys()]

# construct dictionary to keep inferences across 100 iterations
for axe in axes_of_change: 
    dictionary_of_probs_params[axe] = np.zeros((iterations, len(free_params)))

for i in range(iterations): 
    
    grp_sm, dictionary_of_noise_covar, _ = fit_group_shm_with_noise_and_continuous_noise_covariance(dict_of_signals,
                                                                                                 substrates,
                                                                                                 dictionary_of_axes,
                                                                                                 sm_names,
                                                                                                 bvals,
                                                                                                 bvecs,
                                                                                                 shm_degree,
                                                                                                 SNR,
                                                                                                 iterations=100,
                                                                                                 noise_type=noise_type)        
            
    #infer changes across all substrate
    baseline_y, delta_ys = continuous_change(grp_sm, dictionary_of_axes)
    
    for axe in axes_of_change: #iterate over all change axes and compute the inference for this iteration
        
        delta_y = delta_ys[axe]
        sigma_n = dictionary_of_noise_covar[axe]
        dictionary_of_probs_params[axe][i,:], inf_ch, inf_am, _  = ch_mdl.infer(baseline_y, delta_y, sigma_n, parallel=False)
        
plot_probs(dictionary_of_probs_params["axon"].sum(axis=0)/iterations, free_params, title="Axon", fontsize=15)
plot_probs(dictionary_of_probs_params["sphere"].sum(axis=0)/iterations, free_params, title="Soma", fontsize=15)
```
