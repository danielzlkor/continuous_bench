```python
mask = "mask_3"
dwi = np.load("/Users/dkor/OneDrive - Nexus365/BigMac/Analysis/DWdata/self_normalised/{}/dwi.npz".format(mask))["a"]

sms = summary_measures.fit_shm(dwi, shm_degree=2, bval=bvalues, bvec=bvecs)

nissl = np.load("/Users/dkor/OneDrive - Nexus365/BigMac/Analysis/microscopy/{}/SAFs.npz".format(mask))["a"]
myelin = np.load("/Users/dkor/OneDrive - Nexus365/BigMac/Analysis/microscopy/{}/SAFs.npz".format(mask))["b"]
```

```python
dictionary_of_axes = {}
dictionary_of_axes["nissl"] = (nissl/65535)
dictionary_of_axes["myelin"] = (myelin/65535)


#effect_size_nissl = np.percentile(dictionary_of_axes["nissl"],99.9)- np.percentile(dictionary_of_axes["nissl"],0.1)
#effect_size_myelin = np.percentile(dictionary_of_axes["myelin"],99.9)- np.percentile(dictionary_of_axes["myelin"],0.1)

effect_size_nissl = 1
effect_size_myelin = 1



print("effect_size (nissl): {}".format(effect_size_nissl))
print("effect_size (myelin): {}".format(effect_size_myelin))


effect_size_dict = {}

effect_size_dict["baseline"] = 1
effect_size_dict["nissl"] = effect_size_nissl
effect_size_dict["myelin"] = effect_size_myelin
```

#### Set up GLM 

```python
y, X = continuous_change_glm.set_up_GLM(sms, dictionary_of_axes)
y.shape, X.shape
```

```python
beta_dict, covar_dict = continuous_change_glm.get_analytical_betas_and_noise_covariance(y, X, effect_size_dict)
```

```python
## Patterns of change + comparisons 

mus, sigmas = ch_mdl.estimated_change_vectors(beta_dict["baseline"])
plt.figure(figsize=(18, 12))

for i, (m, s) in enumerate(zip(mus, sigmas)):
    plt.subplot(4, 4, i+1)
    continuous_change_glm.plot_changes(sm_names, m , np.diag(s) )
    plt.xticks(fontsize=4)
    
    #if i in [1]:
    #       plt.ylim([0,0.5])
    #else:
    #    plt.ylim([-2.5,2.5])
    
    plt.title(free_params[i])
    
    
plt.subplot(4, 4, 14)
param = "nissl"
continuous_change_glm.plot_changes(sm_names, beta_dict[param], error=np.diag(covar_dict[param]))
plt.title(param, fontsize=15)
plt.xticks(fontsize=4)

plt.subplot(4, 4, 15)
param = "myelin"
continuous_change_glm.plot_changes(sm_names, beta_dict[param], error=np.diag(covar_dict[param]))
plt.title(param, fontsize=15)
plt.xticks(fontsize=4)
```

```python
axes_of_change = [x for x in dictionary_of_axes.keys()]

for axe in axes_of_change:

    prob, _, _, _  = ch_mdl.infer(beta_dict["baseline"],
                                  beta_dict[axe],
                                  covar_dict[axe],
                                  parallel=False)


    continuous_change_glm.plot_probs(probs=prob, free_params=free_params, title=axe, ymax=1)
```
