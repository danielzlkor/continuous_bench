```python
def robust_linear_regression(X,y, huber_epsilon):
    
    import statsmodels.api as sm
    
    rlm = sm.RLM(y, X, M=sm.robust.norms.HuberT(huber_epsilon))
    rlm_fit = rlm.fit()

    # perform a weighted fit with fitted weights to determine R squared
    wls = sm.WLS(rlm.endog, rlm.exog, weights=rlm_fit.weights).fit()
    rlm_r_value = np.sqrt(wls.rsquared)
    rlm_std_err = np.sqrt(wls.bse)
    
    results = {"c" : rlm_fit.params[0],
               "b" : rlm_fit.params[1],
               "p" : rlm_fit.pvalues[-1],
               "r" : rlm_r_value,
               "std" : rlm_std_err,
              "weights": rlm_fit.weights}
    
    return results
```

## Three solutions for this

- Smaller sample size + high iterations ==> to reflect the large residuals of datapoints from the mean. 
-- The large residuals generally mean that it is a very lousy fit.
-- A sign of a lousy fit is that when we're fitting subsets of datapoints, we get very different slopes, which is reflected in our noise covariance term.
-- The point of bootstrapping is to guard against outliers. Here, there is an issue of most point being outliers (i.e., systematically). So, we want smaller sizes as to not include most of them. We want it to actually be reflected as a large covariance.
-- many small but good fits vs. fewer big but bad fits.

- De-weighted the betas after regression with a residual value.
-- Do it once, derive the residuals, and use it to downweight the betas with higher residuals

- Don't use the betas where p>0.05
-- Set the delta = 0, and covariance = 1
