## Signal

```python
## === Signal bookkeeping ===

import csv

def read_csv(filename):
    with open(filename, newline='') as f_input:
        return [list(map(float, row)) for row in csv.reader(f_input)]
    
def signal_invivo_loader(signal_csv, spin_num):
    
    if len(read_csv(signal_csv)) == 0:
        print("{} is empty".format(signal_csv))
        return

    signal = np.array(read_csv(signal_csv)[0])
    
    b0 = spin_num #number of spins

    signal_attenuation = signal/b0
    
    signal_attenuation = signal_attenuation.reshape(1,signal.shape[0])
    
    return signal_attenuation

import re

def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]


def add_noise(data, noise_type='gaussian', SNR=1, S0=1):
    """Add noise to signal

    For Gaussian :
            Snoise = S0 + N(0,sigma)
    For Rician :
            Snoise = sqrt( Real^2 + Imag^2  )
            where Real = S0+N(0,sigma) and Imag=N(0,sigma)
    In both case SNR is defined as SNR = S0/sigma

    Args:
        data : array-like
        noise_type : str
                     accepted values are 'gaussian', 'rician', or 'none'
        SNR : float
        S0  : float
    """
    if noise_type not in ['none','gaussian','rician']:
        raise(Exception(f'Unknown noiise type {noise_type}'))
    if noise_type == 'none':
        return data+0.
    else:
        sigma = S0 / SNR
        if noise_type == 'gaussian':
            noise = np.random.normal(loc=0.0, scale=sigma, size=data.shape)
            return data + noise
        elif noise_type == 'rician':
            noise_r = np.random.normal(loc=0.0, scale=sigma, size=data.shape)
            noise_i = np.random.normal(loc=0.0, scale=sigma, size=data.shape)
            return np.sqrt( (data+noise_r)**2 + noise_i**2 )
```

```python
# === Get dictionary of signals === #

import glob
from os.path import join, basename
from re import split

dict_of_signals = {}
dict_of_noisy_signals = {}
 
axon_list = []
sphere_list = []
spin_number = 160000

    
#rootpath = "/Users/dkor/OneDrive - Nexus365/simulations/BigMac/signals/invivo_3shells/ISMRM_small/signal/160000/5000/1/2.0" #ISMRM substrates simulated with 3-shells, D=2
rootpath = "/Users/dkor/OneDrive - Nexus365/simulations/BigMac/signals/invivo_3shells/ISMRM_large/160000/5000/1" #ISMRM large substrates simulated with 3-shells D=1.7

removed_axon = [290]
removed_spheres = [85, 168]


csv_paths = glob.glob(join(rootpath, "*.csv"))
csv_paths.sort(key=natural_keys)

for i, csv_path in enumerate(csv_paths):
    
    filestrings = split('_',basename(csv_path))
    sphere_num = split('.csv',basename(filestrings[3]))[0]
    
    signal = signal_invivo_loader(csv_path, spin_number)
    
    if signal is not None:
        if (int(filestrings[1]) not in removed_axon) and (int(sphere_num) not in removed_spheres):
            print("Adding axon_{}_sphere_{}".format(filestrings[1], sphere_num))
            dict_of_signals["axon_{}_sphere_{}".format(filestrings[1], sphere_num)] = signal 

            #add noise to signal, and then normalise again:

            noisy_signal = add_noise(signal, noise_type="rician", SNR=100)
            b0_values = noisy_signal[:,(bvals==0)]
            mean_b0 = b0_values.ravel().mean() #mean b0 across all signals

            mean_b0_array = np.repeat(np.array([mean_b0])[...,np.newaxis], bvals.shape[0], axis=1)
            mean_b0_array = np.repeat(mean_b0_array, signal.shape[0], axis=0)
            normalised_signal = noisy_signal/mean_b0_array

            dict_of_noisy_signals["axon_{}_sphere_{}".format(filestrings[1], sphere_num)] = normalised_signal 

            axon_list.append(int(filestrings[1]))
            sphere_list.append(int(sphere_num))
        
sphere_list = np.unique(sphere_list)
axon_list = np.unique(axon_list)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_changes_nicely(names, delta, err, title, xticksize, color, edgecolor,figsize, remove_bottom=False):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])
    
    # Calculate the maximum y-axis limit based on delta, err, and their signs
    ymax = np.max(np.abs(delta + err)) * 1.5
    
    # Set the y-axis limits symmetrically
    ax.set_ylim(-ymax, ymax)

    ax.set_title(title, fontsize=15)
    ax.spines["bottom"].set_visible(True)
    
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    #ax.spines["left"].set_visible(False)

    ax.tick_params(axis='x', which='major', labelsize=xticksize)
    ax.tick_params(axis='y', which='major', size=0)
    
    bars = ax.bar(x=names, height=np.squeeze(delta), yerr=np.squeeze(err), align='center', alpha=1,
           edgecolor=edgecolor, color=color, capsize=10,
                 error_kw=dict(capsize = 10, capthick=5, linewidth=5))
    
    for bar in bars:
        bar.set_linewidth(5)
    #if remove_bottom:
        
    

```

```python
def estimated_change_vectors(ch_mdl, data):
        """
            Returns the change vectors given the baseline measurement (mu, sigma)
        :param data: un-normalized baseline measurements.(n, d) n= number of samples, d= number of measurements
        :return: mu (n, m, d) , sigma_p(n, m, d,d) m=number of change vecs.
        """
        data = np.atleast_2d(data)
        n_samples, d = data.shape
        
        print(len(ch_mdl.summary_names))
        
        assert d == len(ch_mdl.summary_names)
        n_models = len(ch_mdl.models)

        mu = np.zeros((n_samples, n_models, d))
        sigma_p = np.zeros((n_samples, n_models, d, d))

        for i, y in enumerate(data):
            for j, mdl in enumerate(ch_mdl.models):
                #print(mdl.distribution(y))
        
                if j==0:
                    mu[i, j] = ch_mdl.models[j].distribution(data)[0][:,1:]
                    sigma_p[i,j] = ch_mdl.models[j].distribution(data)[1][:,1:,1:]
                else:
                    mu[i, j], sigma_p[i, j] = mdl.distribution(y)
            
                #mu[i, j] = ch_mdl.models[j].distribution(data)[0][:,1:]
                #sigma_p[i,j] = ch_mdl.models[j].distribution(data)[1][:,1:,1:]

        
        return np.squeeze(mu), np.squeeze(sigma_p)
```
